# Reinforcement Learning
The goal is to create an RL network that can balance pole in the CartPole game for 400 frames.

## Summary

In this project two approaches were taken to solve the problem, as shown in the code below:

1.   Using **Q-Learning** led to relatively quicker results, but in spite of much experimentation with hyperparameters, the cart pole was barely able to stay balanced for more than **100 steps** in some of the 20,000 episodes used for training

2.   On the other hand, using **Deep Q-Learning** took much longer to produce results, but the objective was achieved: the cart pole stayed balance for **400 steps** in many of the 600 episodes used for training

## Using Q-Learning

In this approach, the agent is allowed to play and the resulting estimates of the Q-Values are stored in a Q-table as they gradually improve.

```python
# Import packages
import gym
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt; plt.style.use('seaborn-whitegrid')
import matplotlib.animation as animation
mpl.rc('animation', html='jshtml')
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

### Create game environment

```python
# Create environment
env = gym.make('CartPole-v0')
env._max_episode_steps = 400
```

```python
# Initialize environment
env.seed(75)
env.reset()
```

### Set up hyperparameters and Q-table

The following adjustments were made to train the agent and populate the Q-table

- An epsilon-greedy exploration policy was used to alternate between random actions and optimal ones
- The decay rate of the epsilon factor was accelerated to reduce randomness during learning faster
- The size of the Q-table was also reduced so that it could be populated with optimal learned values faster
- The number of episodes was set to 20,000, given the slow convergence of the Q-learning algorithm

```python
# Set up hyperparameters
lr = 0.01
discount = 0.95
n_episodes = 20000
Q_t_size = [15] * len(env.observation_space.high)
window_size = (env.observation_space.high - env.observation_space.low) / Q_t_size
```

```python
# Set up value and decay for epsilon-greedy exploration policy
epsilon = 1
start_decay = 1
end_decay = n_episodes // 3
eps_decay_value = epsilon / (end_decay - start_decay)
```

```python
# Initialize Q-table
q_table = np.random.uniform(low=-2, high=0, \
                            size=(Q_t_size + [env.action_space.n]))

q_table.shape
```

```python
# Create function to generate discrete state
def get_discrete_state(state):
    discrete_state = (state - env.observation_space.low) / window_size
    return tuple(discrete_state.astype(np.int))
```

### Populate Q-table with optimal values

Because the cart pole game gives a reward of +1 for any step taken, then tracking either the sum of rewards or the sum of steps is the same thing, and it allows us to see how far the agent made it before ending the episode.

```python
# Train agent by populating Q-table with optimal values
env.seed(75)
np.random.seed(75)

total_rewards = []

for episode in range(n_episodes):
    
    discrete_state = get_discrete_state(env.reset())
    done = False
    episode_rewards = 0
    
    # Loop until done
    while not done:

        # Use epsilon-greedy exploration policy
        if np.random.random() < epsilon:
            action = np.random.randint(0, env.action_space.n)
        else:
            action = np.argmax(q_table[discrete_state])
        
        # Get new values after taking selected action
        new_state, reward, done, info = env.step(action)
        new_discrete_state = get_discrete_state(new_state)
        
        # Update Q-table values
        next_value = np.max(q_table[new_discrete_state])
        q_table[discrete_state + (action,)] *= (1 - lr)
        q_table[discrete_state + (action,)] += lr * (reward + discount * next_value)
        
        discrete_state = new_discrete_state
        episode_rewards += reward
        
    # Add total episode rewards to list
    total_rewards.append(episode_rewards)

    # Track progress of algorithm
    print("\rEpisode: {}, Steps: {}, eps: {:.3f}".format(episode + 1, episode_rewards, epsilon), end="")
    
    # Implement decay of epsilon value
    if end_decay >= episode >= start_decay:
        epsilon -= eps_decay_value
```

```python
# Print out results
print('Number of episodes: {}'.format(len(total_rewards)))
print('Max steps per episode: {:.0f}'.format(np.max(total_rewards)))
print('Avg steps per episode: {:.0f}'.format(np.mean(total_rewards)))
print('Min steps per episode: {:.0f}'.format(np.min(total_rewards)))
```

```python
# Visualize results
_ = plt.figure(figsize=(8, 4))
_ = plt.plot(total_rewards)
_ = plt.xlabel("Episode", fontsize=14)
_ = plt.ylabel("Steps per episode", fontsize=14)
```

### Test agent using finalized Q-table

```python
# Test and visualize agent
env.seed(75)

frames = []
stop = 0

discrete_state = get_discrete_state(env.reset())
done = False

for step in range(400):
    img = env.render(mode="rgb_array")
    action = np.argmax(q_table[discrete_state])
    new_state, reward, done, info = env.step(action)    
    frames.append(img)
    
    # For the animation we keep printing frames for 5 steps
    stop += done
    if stop > 5:
        break

    discrete_state = get_discrete_state(new_state)
    
print('Number of frames: {:.0f}'.format(len(frames)))
```

```python
# Create functions to update scene and display animation
def update_scene(num, frames, patch):
    patch.set_data(frames[num])
    return patch,

def plot_animation(frames, repeat=False, interval=20):
    fig = plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')
    anim = animation.FuncAnimation(
        fig, update_scene, fargs=(frames, patch),
        frames=len(frames), repeat=repeat, interval=interval)
    plt.close()
    return anim
```

```python
# Visualize animation
plot_animation(frames)
```

## Using Deep Q-Learning

In this approach, instead of using a Q-table to inform the actions of the agent, a neural network is used to take a state and then produce an approximate Q-Value for each possible action.

```python
# Import packages
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.backend import clear_session
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import mean_squared_error
from collections import deque
```

### Create game environment and model

```python
# Create environment
env = gym.make("CartPole-v1")
env._max_episode_steps = 400
input_shape = env.observation_space.shape
n_outputs = env.action_space.n

print('Input shape: {}'.format(input_shape))
print('No. of outputs: {}'.format(n_outputs))
```

```python
# Create deep Q-Network
clear_session()
tf.random.set_seed(75)
np.random.seed(75)

model = keras.models.Sequential([
    keras.layers.Dense(32, activation="elu", input_shape=input_shape),
    keras.layers.Dense(32, activation="elu"),
    keras.layers.Dense(n_outputs)
])
```

### Define exploration policy and hyperparameters

```python
# Define epsilon-greedy exploration policy
def epsilon_greedy_policy(state, epsilon=0):
    if np.random.rand() < epsilon:
        return np.random.randint(n_outputs)
    else:
        Q_values = model.predict(state[np.newaxis])
        return np.argmax(Q_values[0])
```

A buffer is created to store the experiences of the model. In this way, the model will not have to rely only on the latest experiences, but will be able to sample a random batch of past experiences at each training iteration,

```python
# Create buffer to sample past experiences
replay_memory = deque(maxlen=2000)
```

```python
# Create function that samples experiences from buffer
def sample_experiences(batch_size):
    indices = np.random.randint(len(replay_memory), size=batch_size)
    batch = [replay_memory[index] for index in indices]
    states, actions, rewards, next_states, dones = [
        np.array([experience[field_index] for experience in batch])
        for field_index in range(5)]
    return states, actions, rewards, next_states, dones
```

```python
# Create function that takes one step and stores experience
def play_one_step(env, state, epsilon):
    action = epsilon_greedy_policy(state, epsilon)
    next_state, reward, done, info = env.step(action)
    replay_memory.append((state, action, reward, next_state, done))
    return next_state, reward, done, info
```

```python
# Define hyperparameters
batch_size = 32
discount_rate = 0.95
n_episodes = 600
optimizer = Adam(learning_rate=1e-2)
loss_fn = mean_squared_error
```

```python
# Create function that takes training step
def training_step(batch_size):
    experiences = sample_experiences(batch_size)
    states, actions, rewards, next_states, dones = experiences
    next_Q_values = model.predict(next_states)
    max_next_Q_values = np.max(next_Q_values, axis=1)
    target_Q_values = (rewards +
                       (1 - dones) * discount_rate * max_next_Q_values)
    target_Q_values = target_Q_values.reshape(-1, 1)
    mask = tf.one_hot(actions, n_outputs)
    with tf.GradientTape() as tape:
        all_Q_values = model(states)
        Q_values = tf.reduce_sum(all_Q_values * mask, axis=1, keepdims=True)
        loss = tf.reduce_mean(loss_fn(target_Q_values, Q_values))
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
```

### Train model

```python
# Train model
env.seed(75)
np.random.seed(75)
tf.random.set_seed(75)

total_rewards = [] 
best_score = 0

for episode in range(n_episodes):
    obs = env.reset()
    episode_rewards = 0
    for step in range(env._max_episode_steps):
        epsilon = max(1 - episode / int(n_episodes * 0.8), 0.01)
        obs, reward, done, info = play_one_step(env, obs, epsilon)
        episode_rewards += reward
        if done:
            break
    total_rewards.append(episode_rewards)
    if step >= best_score:
        best_weights = model.get_weights()
        best_score = step
    print("\rEpisode: {}, Steps: {}, eps: {:.3f}".format(episode + 1, step + 1, epsilon), end="")
    if episode > 50:
        training_step(batch_size)

model.set_weights(best_weights)
```

```python
# Print out results
print('Number of episodes: {}'.format(len(total_rewards)))
print('Max steps per episode: {:.0f}'.format(np.max(total_rewards)))
print('Avg steps per episode: {:.0f}'.format(np.mean(total_rewards)))
print('Min steps per episode: {:.0f}'.format(np.min(total_rewards)))
```

When the full set of steps taken per episode is visualized, it can be seen that the algorithm performed  progressively better, as the epsilon value used in the exploration policy decayed over time. 

```python
# Visualize results
_ = plt.figure(figsize=(8, 4))
_ = plt.plot(total_rewards)
_ = plt.xlabel("Episode", fontsize=14)
_ = plt.ylabel("Steps per episode", fontsize=14)
```
